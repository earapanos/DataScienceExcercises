#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
    Dica:
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis,
        3. conte o número de categorias dessa variável
        4. acumule essa informação de 3. na lista que você criou em 1.
        5. No final, essa lista pode ser atribuída à nossa variável.

````python
# criando uma variável com o nome das colunas
variaveis = df.columns.tolist()
variaveis
````

````python
# criando um dataframe a partir da lista
df_metadados = pd.DataFrame(variaveis, columns=['variavel'])
df_metadados
````

````python
# lista para armazenar o número de categorias
qtd_categorias = []

# iterar sobre as colunas do dataframe
for coluna in df.columns:
    num_categorias = df[coluna].nunique()
    qtd_categorias.append(num_categorias)
qtd_categorias
````

````python
# adicionando a lista como uma nova coluna no df
df_metadados['qtd_categorias'] = qtd_categorias
df_metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# printando o df
df.head(5)
````

````python
# transformando em dummie
df_encoded  = pd.get_dummies(df, columns=['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'],  drop_first=True, dtype=int)
df_encoded
````

````python
df_encoded.dtypes
````

````python
# transformando categóricas em numéricas
df_encoded.sexo.unique()
````

````python
# transformando categóricas em numéricas
df_encoded['sexo_num'] = df_encoded['sexo'].map({'M': 1, 'F': 0})
df_encoded
````

````python
# transformando categóricas em numéricas
df_encoded.posse_de_veiculo.unique()
````

````python
# transformando categóricas em numéricas
df_encoded['posse_de_veiculo_num'] = df_encoded['posse_de_veiculo'].map({'Y': 1, 'N': 0})
df_encoded
````

````python
# transformando categóricas em numéricas
df_encoded.posse_de_imovel.unique()
````

````python
# transformando categóricas em numéricas
df_encoded['posse_de_imovel_num'] = df_encoded['posse_de_imovel'].map({'Y': 1, 'N': 0})
df_encoded
````

````python
# entendendo o df, iremos excluir as colunas object
df_encoded.info()
````

````python
# criando uma lista com colunas a serem dropadas
columns = ['sexo', 'posse_de_veiculo',  'posse_de_imovel']

df_encoded = df_encoded.drop(columns=columns).reset_index(drop=True)

df_encoded
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# tabela cruzada para a variável possui email e mau
cross_tab_email  = pd.crosstab(df_encoded['possui_email'], df_encoded['mau'], normalize='index')
print("Tabela Cruzada - Possui Email:")
print(cross_tab_email)
````

````python
# tabela cruzada para a variável posse_de_veiculo e mau
cross_tab_veiculo = pd.crosstab(df_encoded['posse_de_veiculo_num'], df_encoded['mau'], normalize='index')
print("\nTabela Cruzada - Posse de Veículo:")
print(cross_tab_veiculo)
````

#### qual a variável mais poderosa?

Observado as proporções entre os resultados das crosstabs entre as variáveis ``possui_email`` e ``posse_de_veiculo`` quando a variável ``mau = 1`` podemos inferir que a variável ``posse_de_veiculo`` na condição apresentada, possui maior poder para definir a variável ``mau = 1``, ou seja, quando o cliente é mau pagador.
