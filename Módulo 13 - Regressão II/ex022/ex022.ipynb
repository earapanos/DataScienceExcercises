```python
# criando a fórmula para o modelo de regressão incluindo todas as variáveis explicativas
y,X = patsy.dmatrices('np.log(renda) ~ C(sexo) + C(posse_de_veiculo) + C(posse_de_imovel) + qtd_filhos + C(tipo_renda) + C(educacao) + C(estado_civil) + C(tipo_residencia) + idade + tempo_emprego + qt_pessoas_residencia',df)
```

```python
# ajustando e exibindo o modelo
modelo = sm.OLS(y, X).fit()

modelo.summary()
```

```python
y, X = patsy.dmatrices('np.log(renda) ~ C(sexo) + C(posse_de_veiculo) + C(posse_de_imovel) + qtd_filhos  + idade + tempo_emprego', df)

# ajustando e exibindo o modelo
modelo = sm.OLS(y, X).fit()

modelo.summary()
```

O modelo final simplificado tem menos variáveis, todas significantes, com uma pequena redução no R-squared (0.347 vs. 0.357), o que sugere que a simplicidade foi alcançada sem perda substancial de explicação.
