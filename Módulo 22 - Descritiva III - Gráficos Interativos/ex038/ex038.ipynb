#### 0. Import das bibliotecas

````python
# Importação das bibliotecas
import pandas as pd
import plotly.express as px
from plotly import graph_objects
from plotly.graph_objs import Layout
````

#### 1. Carregue os dados da tarefa anterior

````python
df = pd.read_csv(
    'https://raw.githubusercontent.com/earapanos/DataScienceExcercises/refs/heads/main/M%C3%B3dulo%2022%20-%20Descritiva%20III%20-%20Gr%C3%A1ficos%20Interativos/ex037/amplitude_diaria_cripto.csv',
    header=[0,1], 
    index_col=0
)
````

````python
# Verificação da estrutura dos dados
print("Estrutura do DataFrame:")
print(df.columns)
print("\nPrimeiras linhas do DataFrame:")
print(df.head())
````

#### 2. Média móvel

1. Faça uma **média móvel**  em 90 dias dos valores de **fechamento ajustados** dos ativos que você escolheu

````python
# Selecionando apenas os preços de fechamento
df_close = df.xs('Close', level=1, axis=1)
````

````python
# Calculando média móvel de 90 dias
df_ma = df_close.rolling(window=90).mean()
print("\nMédia móvel de 90 dias:")
print(df_ma.head(10))
````

````python
# Calculando desvio padrão móvel de 90 dias
df_std = df_close.rolling(window=90).std()
print("\nDesvio padrão móvel de 90 dias:")
print(df_std.head(10))
````

2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

````python
# Preparando dados para plotagem
df_ma_plot = df_ma.stack().reset_index()
df_ma_plot.columns = ['data', 'ativo', 'valor']

df_std_plot = df_std.stack().reset_index()
df_std_plot.columns = ['data', 'ativo', 'valor']
````

3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
# Gráfico de Médias Móveis
fig_ma = px.line(
    data_frame=df_ma_plot.dropna(),
    x='data',
    y='valor',
    color='ativo',
    title='Média Móvel de 90 dias dos Preços de Fechamento',
    labels={'valor': 'Preço (USD)', 'data': 'Data'},
    template='plotly_white'
)
fig_ma.show()
````

````python
# Gráfico de Desvios Padrão
fig_std = px.line(
    data_frame=df_std_plot.dropna(),
    x='data',
    y='valor',
    color='ativo',
    title='Desvio Padrão Móvel de 90 dias dos Preços de Fechamento',
    labels={'valor': 'Desvio Padrão', 'data': 'Data'},
    template='plotly_white'
)
fig_std.show()
````

#### 3. Gráfico de *candlestick*

1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base

````python
# Selecionando dados do BTC-USD para os últimos 60 dias - FORMA CORRETA
# Primeiro precisamos acessar o nível correto do MultiIndex
df_last60 = df['BTC-USD'].tail(n=60)  # Acessa todas as colunas do BTC-USD
print("\nDados dos últimos 60 dias do BTC-USD:")
print(df_last60.head())

# Configuração do layout
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='white',
    height=600,
    title='Gráfico de Candlestick - BTC-USD (Últimos 60 dias)'
)
````

````python
# Dados para o candlestick
graph = {
    'x': df_last60.index,
    'open': df_last60['Open'],
    'close': df_last60['Close'],
    'high': df_last60['High'],
    'low': df_last60['Low'],
    'type': 'candlestick',
    'name': 'BTC-USD',
    'showlegend': True
}

# Criação do gráfico
grafico = graph_objects.Figure(data=[graph], layout=layout)

# Ajustes finais
grafico.update_xaxes(
    showgrid=True,
    gridwidth=1,
    gridcolor='LightGrey'
)
grafico.update_yaxes(
    showgrid=True,
    gridwidth=1,
    gridcolor='LightGrey'
)

grafico.show()
````
