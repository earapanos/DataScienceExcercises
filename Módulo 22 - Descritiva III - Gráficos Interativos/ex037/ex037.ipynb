#### 0. Import das bibliotecas

```python
import pandas as pd
import yfinance as yf
import plotly.express as px
import datetime
import pandas_datareader as pdr
```

#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
ativos = ['BTC-USD', 'ETH-USD', 'USDT-USD']
inicio = '2020-01-01'
fim = None  # ou use uma data final como '2024-12-31'
```

```python
# Baixar todos os ativos de uma vez
df = yf.download(ativos, start=inicio, end=fim, group_by='ticker')

df.head()
```

#### 2. Observando a volatilidade dentro do dia

1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).

```python
for ticker in ativos:
    df[ticker, 'Price Range'] = df[ticker]['High'] - df[ticker]['Low']
```

```python
df_range = df.xs('Price Range', axis=1, level=1).stack().reset_index()
df_range.columns = ['data', 'ativo', 'valor']
```

2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.

```python
fig = px.line(data_frame=df_range, 
              x='data', 
              y='valor', 
              color='ativo', 
              title='Amplitude Diária de Preço (High - Low) das Criptomoedas',
              template='plotly_white',
              labels={'data': 'Data', 'valor': 'Amplitude (USD)', 'ativo': 'Ativo'})

fig.show()
```

3. Veja se consegue identificar que ativo apresentou maior volatilidade.

> Entre as três principais criptomoedas selecionadas como ativos para esta atividade, podemos observar claramente que o Bitcoin USD (**BTC-USD**) apresentou a maior volatilidade durante o período de análise desde o ano de 2020.
