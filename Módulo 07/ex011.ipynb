#### 2) Vamos para o modelo:

1. Defina um objeto com a função da árvore de decisão vista em aula.
2. Treine o modelo com os dados que você separou para treinamento.
3. Visualize a árvore. Talvez você queira aumentar um pouco a figura.
4. Produza uma visualização da matriz de classificação (ou matriz de confusão) - coloque os rótulos como "aprovados" e "reprovados" - pois afinal de contas, são essas as decisões que se podem tomar com propostas de crédito.
5. Calcule a acurácia na base de treinamento

````python
# criando o objeto árvore através da variável classifier
clf = DecisionTreeClassifier(random_state=100, 
                             splitter='random')
clf
````

````python
# definindo variáveis de entrada e alvo
X = df.drop(columns=['mau'])
y = df.mau
````

````python
# escrevendo a variável alvo
y.value_counts()
````

````python
# treinando o modelo conforme as variáveis de entrada e alvo
clf = clf.fit(X,y)
clf
````

````python
# plotando a figura da árvore
plt.figure(figsize=(25,10))
plot_tree(clf,
          feature_names=X.columns,
         filled=True,
          proportion=False,
         class_names=['Bom', 'Mau'],
         fontsize=8);
````

````python
# Fazendo previsões usando o modelo treinado
y_pred = clf.predict(X)
````

````python
# calculando a matriz de confusão
cm = confusion_matrix(y, y_pred)
````

````python
# Exibindo a matriz de confusão
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Bom', 'Mau'])
disp.plot()
plt.show()
````

````python
# fazendo previsões usando o modelo treinado
y_pred = clf.predict(X)

# calculando a acurácia
accuracy = accuracy_score(y, y_pred)
````

````python
print(f'A acurácia do modelo é: {accuracy}%')
````

### Separando entre treino e teste

````python
# separando os dados
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=100)
````

````python
# treinando o classificador
clf = DecisionTreeClassifier(random_state=100)
clf.fit(X_train, y_train)
````

````python
# Prevendo com os dados de treino
y_hat_train = clf.predict(X_train)
````

````python
# matriz de coonfusão para os dados de treino
cm_train = confusion_matrix(y_train, y_hat_train, labels=clf.classes_)
MatrizDeConfusao_train = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=['bom', 'mau'])
MatrizDeConfusao_train.plot()
````

````python
# acurácia do modelo nos dados de treino
accuracy_train = accuracy_score(y_train, y_hat_train)
print(f'A acurácia do modelo de treino é: {accuracy_train}%')
````
