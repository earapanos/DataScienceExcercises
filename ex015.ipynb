````python
# importando bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings 

warnings.filterwarnings('ignore')
````

````python
# importando o dataset
df = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
df.head()
````

````python
# entendendo os tipos de variáveis disponíveis
df.dtypes
````

````python
df.data_ref.unique()
````

````python
# entendendo a relação entre sexo e renda
ax = sns.pointplot(x='sexo'
                   , y='renda'
                   , data=df
                   , dodge=True
                   , ci = 90)
````

````python
# entendendo a relação entre sexo e tempo_emprego
ax = sns.pointplot(x='sexo'
                   , y='tempo_emprego'
                   , data=df
                   , dodge=True
                   , ci = 90)
````

````python
# entendendo a relação entre sexo e idade
ax = sns.pointplot(x='sexo'
                   , y='idade'
                   , data=df
                   , dodge=True
                   , ci = 90)
````

````python
# entendendo a relação entre sexo e renda
ax = sns.pointplot(x='data_ref'
                   , y='renda'
                   , data=df
                   , dodge=True
                   , ci = 100)
plt.xticks(rotation=90)  
plt.show()
````

````python
# plotando a relação da quantidade de filhos, renda com a variável de mau pagador
ax = sns.pointplot(x="qtd_filhos", y="renda", hue = 'mau',
                   data=df, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
plt.xticks(rotation=90)  
plt.show()
````
